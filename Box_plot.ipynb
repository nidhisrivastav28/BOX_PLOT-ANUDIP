```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your Excel file into a Pandas DataFrame
excel_file = 'Data Analysis - Caselet#2.xlsx'
sheet_name = 'DifferentFormat'
df = pd.read_excel(excel_file, sheet_name=sheet_name)

# Create a box plot using Seaborn
plt.figure(figsize=(6, 6))  # Adjust the figure size if needed
bp = df.boxplot(column='TPT', by='Team Member', vert=True, patch_artist=True, grid= False)

plt.title('BOX PLOT OF TPT COLUMN FOR EACH STUDENT')
plt.xlabel('Team Member')
plt.ylabel('TPT Values')
plt.xticks(rotation=90)
plt.show()
```
